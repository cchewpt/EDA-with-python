```python
# Showing data by choosing specific columns
df = dataframe[['Movie Name', 'Year of Release','Genre', 'Director',
                'Movie Rating', 'MetaScore', 'Gross', 'Stars']].copy()
```

```python
# Rename column names
df = df.rename(columns = {'Movie Name' : 'movie_name', 
                     'Year of Release' : 'release_year',
                     'Genre' : 'genre', 
                     'Director' : 'director',
                     'Movie Rating' : 'movie_rating', 
                     'MetaScore' : 'metascore',
                     'Gross' : 'gross', 
                     'Stars' : 'stars'
                    })
```

```python
ax = df['release_year'].value_counts()\
     .head(10)\
     .plot(kind = 'bar', title = "Most movie release in a year")
ax.set_xlabel('Year')
ax.set_ylabel('Amount of movies')
```

```python
ax = df['genre'].value_counts()\
           .head(10)\
           .plot(kind = 'barh', title = "Most genre")
ax.set_ylabel('Genre')
ax.set_xlabel('Amount')
```

```python
ax = df['movie_rating'].plot(kind = "hist",
                        bins = 20,
                        title = "Movie rating")
ax.set_ylabel('Amount')
ax.set_xlabel('Rating')
```

```python
 ax = df.head(10000).plot(kind = "scatter",
        x = 'movie_rating', 
        y = 'gross',
        title = 'relationship between rating and gross')
ax.set_xlabel('Rating')
ax.set_ylabel('Gross (billion)')
plt.show()
```
